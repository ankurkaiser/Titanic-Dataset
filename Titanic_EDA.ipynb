Now its time to get rid of outliers. Outliers may affect the prediction of the model later. ##Outliers are datapoints that lie beyond s.d +-3.

```python
def remove_outliers(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR= Q3-Q1
    
    lower_range=Q1-(1.5*IQR)
    higher_range=Q1+(1.5*IQR)
    return lower_range,higher_range

lowage,highage=remove_outliers(data['Age'])

data['Age']=np.where(data['Age']<lowage,lowage,data['Age'])
data['Age']=np.where(data['Age']>highage,highage,data['Age'])
```

```python
remove_outliers(data['Age'])
```
